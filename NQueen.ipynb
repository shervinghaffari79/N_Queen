```python
import random

def generate_indivisual(N_Queen):

    indivisual = list(range(N_Queen))
    random.shuffle(indivisual)
    return indivisual

def generate_population(population_size,N_Queen):

    population = []
    for _ in range(population_size):
        population.append(generate_indivisual(N_Queen))
    return population


def fitness(indivisual):
    Score = 0
    for i in range(len(indivisual)):
        for j in range(i+1,len(indivisual)):
            if indivisual[i] != indivisual[j] and abs(indivisual[i] - indivisual[j]) != abs(i-j):
                Score+=1
    return Score


def selection(population):
    parents = []
    for _ in range(2):
        sample = random.sample(population, k=5)
        sample.sort(key = lambda x: fitness(x))
        parents.append(sample[-1])
    return parents


def crossover(parents):
    e = random.randint(0,len(parents[0]))
    offspring = [None]*len(parents[0])
    flag = 0
    for i in range(e):
        if fitness(parents[0])>fitness(parents[1]):
            offspring[i] = parents[0][i]
            flag = 1
        else:
            offspring[i] = parents[1][i]
            flag = 0
    j = 0
    for index in range(len(offspring)):
        if offspring[index] == None:
            while parents[flag][j] in offspring:
                j+=1
            offspring[index] = parents[flag][j]
            j+=1

    return offspring

def mutation(offspring):

    if random.random()<0.5 :
        i = random.randint(0, len(offspring) - 1)
        j = random.randint(0, len(offspring) - 1)
        offspring [i] ,offspring[j] = offspring[j],offspring[i]
    return offspring

def Evolutionary (N_Queen,generation_size,population_size):

    population = generate_population(population_size, N_Queen)
    new_generation = []
    for _ in range(generation_size):

        for _ in range(population_size):

            parents = selection(population)
            offspring = crossover(parents)
            offspring = mutation(offspring)
            new_generation.append(offspring)
            new_generation.append(parents[0])
            new_generation.append(parents[1])
        population = new_generation

    population.sort(key=lambda x : fitness(x))

    display(population[-1])

    print("fault : " ,int(N_Queen/2*(N_Queen-1))-fitness(population[-1]))

def display(board):
    chess = [['_' for _ in range(len(board))] for _ in range(len(board))]
    for row in range(len(board)):
        for col in range(len(board)):
            if board[col] == row:
                print("1",end=" ")
            else:
                print("_",end=" ")
        print()


Evolutionary(16,100,100)

```
